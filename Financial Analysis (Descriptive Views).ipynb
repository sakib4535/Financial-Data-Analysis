```python
close["MSFT"].shift(periods=1).to_frame()
```

```python
msft = close.MSFT.dropna().to_frame().copy()
```

```python
msft
```

```python
msft.rename(columns={"MSFT":"Price"}, inplace=True)
```

```python
msft
```

```python
msft.shift(periods=1)
```

```python
msft["price_lag1"] = msft.shift(periods=1)
msft
```
